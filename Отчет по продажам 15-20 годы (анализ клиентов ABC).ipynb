Приведение данных в соответствии с ABC анализом

Если суммы продаж не отсортированы от большего к меньшему, то надо отсортировать

```python
df.sort_values('sale', ascending=False)
```

```python
dfsale = df.sort_values('sale', ascending=False) #сохраним отсортированный фрейм в "dfsale"
```

```python
dfsale['dol'] = dfsale['sale']/dfsale['sale'].sum() #вычислим долю суммы отгрузок каждого клиента в сумме общих продаж
```

Посчитаем долю нарастающим итогом.

```python
dfsale['dolpr'] = dfsale['dol']
```

```python
dfsale['dolpr'][1:] = 0
```

```python
for i in range(1,len(dfsale.index)):
    dfsale['dolpr'].iloc[i] = dfsale['dol'].iloc[i] + dfsale['dolpr'].iloc[i-1]
```

```python
dfsale
```

Добавим колонку "группа" и, сначала, назначим всем клиентам группу С

```python
dfsale['group'] = "C"
```

```python
dfsale['group'][dfsale.dolpr <= 0.8] = "A" 
```

```python
dfsale['group'][(dfsale.dolpr <= 0.95) & (dfsale.group != "A")] = "B" 
```
